### <u>Popcorn Hack 1 - Dice Roller</u>


Instructions:
Finish the code in the cell below (fill in underlines and add lines of code) so it works as described:

- Create a class called Dice that:
    - Has a property for the number of sides.
    - Has a method roll() that returns a random number between 1 and the number of sides.


At the bottom, test your game by making a DiceGame object and calling roll() a few times.

The code is currently NOT functional. When you have completed this popcorn hack, the code should work. Have fun!

```python
%%javascript

// Create a Dice class
class Dice {
  constructor(sides) {
    this.sides = sides; // property for number of sides
  }

  // roll() method returns a random integer from 1 to sides
  roll() {
    return Math.floor(Math.random() * this.sides) + 1;
  }
}

// Instantiate a Dice object
let myDice = new Dice(6);

// Test the roll method a few times
console.log("üé≤ Popcorn Hack 1 - Dice Roller üé≤");

// Roll 3 times and log each result
for (let i = 1; i <= 3; i++) {
  let result = myDice.roll();
  console.log(`Roll ${i}: ${result}`);
}
```

```python
%%javascript
// Create or reuse a container element to display the pet UI
(function() {
  let element = document.getElementById('petDisplay');
  if (!element) {
    element = document.createElement('div');
    element.id = 'petDisplay';
    // add minimal styling so it's visible in the notebook
    element.style.border = '1px solid #ddd';
    element.style.padding = '12px';
    element.style.margin = '8px 0';
    element.style.maxWidth = '500px';
    document.body.appendChild(element);
  }

  // Pet class
  class Pet {
    constructor(name) {
      this.name = name;
      // starting stats (you can tweak these)
      this.hunger = 1;      // almost full
      this.energy = 8;      // energetic
      this.happiness = 10;  // very happy
    }

    feed() {
      this.hunger = Math.max(0, this.hunger - 2);
      this.energy = Math.min(10, this.energy + 1);
      this.happiness = Math.min(10, this.happiness + 1);
      this.status(`${this.name} just enjoyed a big snack! üçó`);
    }

    play() {
      this.energy = Math.max(0, this.energy - 2);
      this.hunger = Math.min(10, this.hunger + 1);
      this.happiness = Math.min(10, this.happiness + 1);
      this.status(`${this.name} had a blast playing! üéâ`);
    }

    sleep() {
      this.energy = Math.min(10, this.energy + 3);
      this.hunger = Math.min(10, this.hunger + 1);
      this.status(`${this.name} took a cozy nap. üò¥`);
    }

    // update the DOM display and reattach button handlers
    status(message) {
      element.innerHTML = `
        <h3>${message}</h3>
        <p><strong>${this.name}'s Current Stats</strong></p>
        <ul>
          <li>Hunger: ${this.hunger}</li>
          <li>Energy: ${this.energy}</li>
          <li>Happiness: ${this.happiness}</li>
        </ul>
        <div style="margin-top:8px;">
          <button id="feedBtn">Feed</button>
          <button id="playBtn">Play</button>
          <button id="sleepBtn">Sleep</button>
          <button id="logBtn" style="margin-left:8px;">Log ‚Üí Console</button>
        </div>
      `;

      // attach handlers (use arrow functions to keep `this`)
      document.getElementById('feedBtn').onclick = () => this.feed();
      document.getElementById('playBtn').onclick = () => this.play();
      document.getElementById('sleepBtn').onclick = () => this.sleep();
      document.getElementById('logBtn').onclick = () => this.logToConsole();
    }

    // optional helper to print to console for debugging
    logToConsole() {
      console.log(`--- ${this.name} stats ---`);
      console.log('Hunger:', this.hunger);
      console.log('Energy:', this.energy);
      console.log('Happiness:', this.happiness);
      console.log('-----------------------');
    }
  }

  // Create the pet and show initial status
  // (You asked for "Anish" with high happiness/low hunger ‚Äî set above)
  if (!window._popcornPetInstance) {
    window._popcornPetInstance = new Pet('Anish');
  }
  window._popcornPetInstance.status(`${window._popcornPetInstance.name} is ready! Click a button to interact.`);
})();
```

```python
%%javascript
// ====================
// HOMEWORK CLASSES
// ====================

// 1Ô∏è‚É£ Person Class
class Person {
  constructor(name, age) {
    this.name = name;
    this.age = age;
  }

  greet() {
    console.log(`Hello! My name is ${this.name} and I am ${this.age} years old.`);
  }
}

// Test Person
let person1 = new Person("Alice", 25);
console.log("üë§ Person Class üë§");
console.log(person1);
person1.greet();


// 2Ô∏è‚É£ Book Class
class Book {
  constructor(title, author, pages) {
    this.title = title;
    this.author = author;
    this.pages = pages;
    this.pagesRead = 0;
  }

  read(pages) {
    this.pagesRead += pages;
    if (this.pagesRead > this.pages) this.pagesRead = this.pages;
    console.log(`You have read ${this.pagesRead} pages of "${this.title}".`);
  }

  isLongBook() {
    return this.pages > 300;
  }
}

// Test Book
let book1 = new Book("JavaScript Basics", "John Doe", 350);
console.log("üìö Book Class üìö");
console.log(book1);
book1.read(50);
console.log("Is it a long book?", book1.isLongBook());


// 3Ô∏è‚É£ Car Class
class Car {
  constructor(make, model, fuel) {
    this.make = make;
    this.model = model;
    this.fuel = fuel;
  }

  drive(distance) {
    const fuelUsed = distance * 0.1;
    if (fuelUsed > this.fuel) {
      console.log(`Not enough fuel to drive ${distance} km.`);
      return;
    }
    this.fuel -= fuelUsed;
    console.log(`${this.make} ${this.model} drove ${distance} km. Fuel left: ${this.fuel.toFixed(2)}L`);
  }

  refuel(amount) {
    this.fuel += amount;
    console.log(`${this.make} ${this.model} refueled ${amount}L. Total fuel: ${this.fuel.toFixed(2)}L`);
  }

  static info() {
    console.log("Cars are vehicles.");
  }
}

// Test Car
let car1 = new Car("Toyota", "Corolla", 10);
console.log("üöó Car Class üöó");
car1.drive(50);
car1.refuel(20);
Car.info();


// 4Ô∏è‚É£ ElectricCar Class
class ElectricCar extends Car {
  constructor(make, model, fuel, battery) {
    super(make, model, fuel);
    this.battery = battery;
  }

  charge(amount) {
    this.battery += amount;
    if (this.battery > 100) this.battery = 100;
    console.log(`${this.make} ${this.model} charged. Battery: ${this.battery}%`);
  }
}

// Test ElectricCar
let eCar = new ElectricCar("Tesla", "Model 3", 0, 50);
console.log("‚ö° ElectricCar Class ‚ö°");
eCar.charge(30);
eCar.drive(30);


// 5Ô∏è‚É£ Routine combining all objects
console.log("üîÑ Routine combining all objects üîÑ");

let person2 = new Person("Bob", 30);
let book2 = new Book("Advanced JS", "Jane Smith", 200);
let car2 = new Car("Honda", "Civic", 15);
let eCar2 = new ElectricCar("Tesla", "Model X", 0, 20);

person2.greet();
book2.read(40);
console.log("Is the book long?", book2.isLongBook());
car2.drive(100);
car2.refuel(10);
eCar2.charge(50);
eCar2.drive(30);
```
